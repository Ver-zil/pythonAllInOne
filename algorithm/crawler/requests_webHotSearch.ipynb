# web hot search

```python
import requests
import re
import parsel
import os
import csv
import pandas as pd
```

```python
url = 'https://weibo.com/ajax/side/hotSearch'

headers = {
    'User-Agent':
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
}

save_path = r'D:\cache\coding\python\crawler\\'

# client = httpx.Client(http2=True)
# response = client.get(url, timeout=30)
response = requests.get(url, headers=headers)
```

```python
# f = open(save_path + 'webHotSearch.csv', mode='a', encoding='utf-8')
# csv_writer = csv.DictWriter(f, fieldnames=['serial', 'topic', 'category'])
list = []

for num, hot in enumerate(response.json()['data']['realtime']):
    if 'category' not in hot:
        continue
    dict = {'serial': num, 'topic': hot['word'], 'category': hot['category']}
    # csv_writer.writerow(dict)
    list.append(dict)

pd_data = pd.DataFrame(list)
pd_data.to_csv(save_path + 'webHotSearch.csv')
```
