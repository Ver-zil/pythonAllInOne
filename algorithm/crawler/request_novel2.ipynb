# Download the whole novel

```python
import requests
import re
import parsel
import httpx
import os
import pprint
```

```python
url = 'https://www.biqudu.com/0_390/'

headers = {
    'User-Agent':
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
}

save_path = r'D:\cache\coding\python\crawler\\'

# client = httpx.Client(http2=True)
# response = client.get(url, timeout=30)
response = requests.get(url, headers=headers)
```

```python
selector = parsel.Selector(response.text)
book_name = selector.xpath('//div[contains(@id,"info")]/h1/text()').get()
# chapters_info = selector.xpath('//div[@class="listmain"]//dd/a').getall()[12:]
chapter_urls = list(map(lambda chapter_url: url + chapter_url, re.findall('<dd><a href ="/0_390/(.*?)">' ,response.text)))[12:]
```

```python
# save_path = save_path + book_name
# if not os.path.exists(save_path):
#     os.mkdir(save_path)
```

```python
with open(save_path + book_name + '.txt', mode='a', encoding='utf-8') as f: 
    for c_url in chapter_urls:
        if c_url == chapter_urls[5]:
            break
        chapter_response = requests.get(c_url, headers=headers)
        selector = parsel.Selector(chapter_response.text)
        title = selector.xpath('//div[contains(@class, "reader")]//h1/text()').get()
        content = '\n\n'.join(selector.xpath('//*[contains(@id, "content")]/text()').getall()[:-2])
        f.write(title + '\n\n')
        f.write(content + '\n\n\n\n')
```
