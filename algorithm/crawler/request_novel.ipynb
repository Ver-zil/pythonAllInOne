```python
# xpath和css的匹配方式类似，一般可以互相转换
selector = parsel.Selector(response.text)

# xpath里/和//含义不一样，/表示绝对路径，//相对 可以匹配到其更次级的子标签内的内容
title = selector.xpath('//div[contains(@class, "reader")]//h1/text()').get()
content = '\n\n'.join(
    selector.xpath('//*[contains(@id, "content")]/text()').getall()[:-2])

# title = selector.css('.reader h1::text').get()
# content = '\n'.join(selector.css('#content::text').getall())
```
